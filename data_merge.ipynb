```python
import pandas as pd
```

```python
master_frame = pd.read_csv('preprocessed_data.csv')
```

```python
#Drop unnecessary columns
master_frame.drop(['MEDREIMB_OP', 'BENRES_OP', 'PPPYMT_OP','DEATH',
                   'fst_admsn_dt'], axis=1, inplace=True)
```

```python
#master_frame.drop('CLM_ADMSN_DT',axis=1, inplace=True)
```

```python
#Share of patients who died in the dataset. Because it's less than 2% of the total, we can remove them 
master_frame[master_frame.DEATH_DUMMY==1].shape[0]/master_frame.shape[0]
```

```python
#Maser frame without the patients who died
master_frame=master_frame[master_frame.DEATH_DUMMY==0]
```

```python
master_frame.drop('DEATH_DUMMY',axis=1,inplace=True)
```

```python
master_frame.head()
```

```python
#Read the RX file
rx = pd.read_csv('RX.csv')
```

```python
rx.head()
```

```python
#Merge with RX 
master_frame = pd.merge(master_frame, rx[['DESYNPUF_ID','#Rx','$Rx']], 
                        how='left', on='DESYNPUF_ID')
```

```python
master_frame.head()
```

```python
#Read the Carrier file
carrier = pd.read_csv('carrier.csv')
```

```python
carrier.head()
```

```python
#Merge with Carrier data
master_frame = pd.merge(master_frame, carrier[['DESYNPUF_ID','carrier_pmt','carrier_number_claims',
                                               'office_visits', 'office_visits_pmt']], 
                        how='left', on='DESYNPUF_ID')
```

```python
master_frame.head()
```

```python
#Read Outpatient file
op = pd.read_csv('op.csv')
```

```python
op.head()
```

```python
#Merge with Outpatient data
master_frame = pd.merge(master_frame, op[['DESYNPUF_ID','CLM_ID','TOTAL_PAYMENT_PERID']], 
                        how='left', on='DESYNPUF_ID')
```

```python
master_frame.head().transpose()
```

```python
#Add total payment field
master_frame['total_pmt'] = master_frame[['ip_pmt','$Rx', 'carrier_pmt', 
                                     'TOTAL_PAYMENT_PERID']].sum(axis=1)
```

```python
master_frame.head(2).transpose()
```

```python
#Columns to divie by MM
cols = ['unique_ip_dt','ip_pmt','days_frm_last_ip','#Rx','$Rx', 
        'carrier_pmt', 'carrier_number_claims', 'office_visits', 'office_visits_pmt', 'CLM_ID',
       'TOTAL_PAYMENT_PERID', 'total_pmt']
```

```python
#Add columns divided by MM
for col in cols:
    master_frame[str(col) + '_mm'] = master_frame[str(col)]/master_frame.member_months
```

```python
#Transpose for checking math
master_frame.head(1).transpose()
```

```python
master_frame.head()
```

```python
#Write pre-selected data into a csv for records
master_frame.to_csv('master_frame_all_records.csv')
```

```python
master_frame.columns
```

```python
master_model = master_frame[['DESYNPUF_ID', 'BENE_SEX_IDENT_CD', 'BENE_RACE_CD', 'SP_STATE_CODE', 
                        'SP_ALZHDMTA', 'SP_CHF', 'SP_CHRNKIDN', 'SP_CNCR', 'SP_COPD',
       'SP_DEPRESSN', 'SP_DIABETES', 'SP_ISCHMCHT', 'SP_OSTEOPRS', 'SP_RA_OA',
       'SP_STRKETIA', 'Y', 'TRUE_AGE', 'unique_ip_dt_mm', 'ip_pmt_mm',
       'days_frm_last_ip_mm', '#Rx_mm', '$Rx_mm', 'carrier_pmt_mm',
       'carrier_number_claims_mm', 'office_visits_mm', 'office_visits_pmt_mm',
       'CLM_ID_mm', 'TOTAL_PAYMENT_PERID_mm', 'total_pmt_mm' ]]
```

```python
#Bucketing age model
def bucket_age(age):
    if age < 65:
        return 1
    elif age < 75:
        return 2
    elif age < 85:
        return 3
    else:
        return 4
```

```python
master_model['age_bucket'] = master_model['TRUE_AGE'].apply(lambda x: bucket_age(x))
```

```python
master_model.head().transpose()
```

```python
master_model.drop('TRUE_AGE',axis=1,inplace=True)
```

```python
master_model.head().transpose()
```

```python
#Check for NAN
master_model.columns[pd.isnull(master_model).sum()>0].tolist()
```

```python
#Fill missing values with 0
master_model.fillna(0, inplace = True)
```

```python
#Write  data into a csv for records
master_model.to_csv('master_model_data.csv')
```

```python
Y = master_model.pop("Y")
```
